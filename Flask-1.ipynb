Q3. What is App routing in Flask? Why do we use app routes?

Ans. App routing in flask means mapping a unique URL to all the functions defined in our code. App routing is done using the decorator **@app.route("/")**, the url is provided in the parenteses as a string, for example ("/url").

App route help us in deciding which urls can be used to access particular functions of our app. When a user accesses a url, flask executes the corresponding function and gives the relevant html result.

App routes help in making our application structured.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

Ans. The **url_for()** function of Flask helps in url building. It can dynamically build url based on our app routes. 
Use of url_for is as show to display various urls:

```python
#Solution-5
#creating a simple app
from flask import url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'My Home Page'

@app.route('/about')
def about():
    return 'My about page'

@app.route('/about/contact')
def contact():
    return 'My contact details page'

@app.route('/profile')
def profile():
    return 'My profile page'

@app.route('/urls')
def urls():
    #obtaining urls and displaying
    
    a = url_for('home')
    b = url_for('about')
    c = url_for('contact')
    d = url_for('profile')
    
    return f"My URLs:<br>URL for home : {a}<br>URL for about : {b}<br>URL for contact : {c}<br>URL for profile : {d}<br>"

if __name__ == "__main__":
    app.run(host="0.0.0.0")
```
