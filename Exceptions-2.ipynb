Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans. LookupError class is used to raise exceptions when an element of an iterable is not found.
1. IndexError- This occurs when the index called is out of range of index of iterable
2. KeyError- it is raised when the specified key doesn't exist in the dictionary

```python
#IndexError
l = [1,2,3]
print(l[7])
```

```python
#KeyError
d = {1:'a', 2:'b', 3:'c'}
print(d[7])
```

Q6. List down some best practices for exception handling in python.
1. Use specific exception classes not generic - for better understanding of anyone reading code 
2. Always use a proper message - for proper debugging
3. Always log the errors - for debugging
4. Avoid to write multiple exception handlings - to reduce coding complexity
5. Document all the errors 
6. Cleanup all the resources using finally in the end - so that resources are not over-utilized. Example, closing a file, shutting down logging
