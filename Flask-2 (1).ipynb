Q1. Explain GET and POST methods.

Ans. Both GET and POST methods are used for passing data through HTTP. But there are differences-

- **GET** - When GET method is used, the data is passed in the form of url, through the address bar. This is good where we don't need to secure our data. For example, when we do a google search, GET method is used.

- **POST** - When POST method is used, the data is passed in the form of body. This is more secured type of data transfer. For example, when we put our email id and password in google login, POST method is used.

Q3. Why is redirect() used in Flask?

Ans. The **redirect()** method is used to redirect the browser to a different url, which is provided as argument to the redirect() method. For example, after a payment, we want to redirect user to payment successful page.

Example is as follows:

```python
from flask import Flask, request, render_template, url_for, redirect

app = Flask(__name__)

@app.route('/')
def home_page():
    return render_template('index.html')

@app.route('/submit',methods=['POST'])
def submit():
    if(request.method == 'POST'):
        
        name = request.form['name']  
        return redirect(url_for('success', name=name))

@app.route('/success/<name>')
def success(name):
    result = f'Thanks for your Submission {name}'
    return render_template('results.html', result= result)

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
